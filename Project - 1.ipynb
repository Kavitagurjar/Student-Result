# Drop unnamed colomn

```python
df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

# Gender Distribution

```python
plt.figure (figsize=(4,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
# From the above chart we have analysed that :
# the number of female in the data is more than the number of males 
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean', "WritingScore":'mean'})
print(gb)
```

```python
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parent's Education and Student's score")
plt.show()
```

```python
#from the above chart we have concluded that the education of the parents have a good impact on their study
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean', "WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1, annot = True)
plt.title("Relationship between Parent's Marital Status and Student's score")
plt.show()
```

```python
# the above chart we have concluded that there is no/negligible impact on the student's score due to their parent's marital status.
```

<h1>Distribution of Ethinic Groups :-</h1>

```python
group_A = df.loc[df["EthnicGroup"] == "group A"].count()
group_B = df.loc[df["EthnicGroup"] == "group B"].count()
group_C = df.loc[df["EthnicGroup"] == "group C"].count()
group_D = df.loc[df["EthnicGroup"] == "group D"].count()
group_E = df.loc[df["EthnicGroup"] == "group E"].count()

l = ["group A","group B","group C","group D","group E"]
group_list = [group_A["EthnicGroup"], group_B["EthnicGroup"], group_C["EthnicGroup"], group_D["EthnicGroup"], group_D["EthnicGroup"]]
print(group_list)

plt.pie(group_list, labels = l, autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```
